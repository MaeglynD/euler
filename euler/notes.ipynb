```python
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#
# prime numbers
#

def prime_gen(n):
  nums = list(range(0, n+1))
  primes = [2]
  curr = 2

  while True:
    old_curr = curr

    for i in range(curr, n+1, curr):
      nums[i] = False

    for i in range(curr+1, n+1):
      if nums[i]:
        primes.append(i)
        curr = i
        break
    
    if old_curr == curr:
      break
  
  return primes

n = 100
primes = prime_gen(n)

np.set_printoptions(threshold=np.inf)
np.set_printoptions(linewidth=np.inf)

data = np.zeros((n, len(primes)))


def matrix_of_prime_factors(data):
  for i in range(2, n):
    if i in primes:
        data[i][primes.index(i)] = 1
        continue

    for j in range(len(primes)):
      tmp = i
      c = 0

      while tmp % primes[j] == 0:
        c +=1
        tmp /= primes[j]

      data[i][j] = int(c)

  return data

data = matrix_of_prime_factors(data)


def plot_heatmap(matrix):
  plt.figure(figsize=(10, 20))
  plt.imshow(matrix, interpolation='nearest')

  for i in range(matrix.shape[0]):
    for j in range(matrix.shape[1]):
      if matrix[i, j] != 0:
        plt.text(j, i, str(int(matrix[i, j])), ha='center', va='center', color='white')

  plt.xticks(range(matrix.shape[1]), [f'{primes[j]}' for j in range(matrix.shape[1])])
  plt.yticks(range(matrix.shape[0]), [f'{i}' for i in range(matrix.shape[0])])
  plt.colorbar(label='Value')
  plt.clim(1, data.max())
  plt.grid(False)
  plt.tight_layout()
  plt.show()

# plot_heatmap(data)


def plot_3d_heatmap(matrix):
  matrix = data[:,:40]
  fig = plt.figure(figsize=(10, 10))
  ax = fig.add_subplot(111, projection='3d')

  xpos, ypos = np.meshgrid(np.arange(matrix.shape[1]), np.arange(matrix.shape[0]))
  xpos = xpos.flatten()
  ypos = ypos.flatten()
  zpos = np.zeros_like(xpos)
  zval = matrix.flatten()

  colors = plt.cm.viridis(zval / float(max(zval)))

  ax.bar3d(xpos, ypos, zpos, 1, 1, zval, color=colors, zsort='average')
  plt.xticks(range(matrix.shape[1]), [f'{primes[j]}' for j in range(matrix.shape[1])])

  plt.show()
  

def print_colored_text(text):
  colored_text = ''.join('\x1b[30;2;50m' + char + '\x1b[0m' if char == '0' else char for char in text)
  print(colored_text)


def test_vis(data):
  test = []

  for row in data:
    tmp = ''
    for val in row:
      if val != 0:
        tmp += str(int(val))
    
    test.append(tmp)

  print_colored_text('\n'.join(test))

# test_vis(data)
```
